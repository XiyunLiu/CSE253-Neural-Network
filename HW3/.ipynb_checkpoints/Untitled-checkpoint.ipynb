```python
import numpy as np
import matplotlib.pyplot as plt
import random
import math
```

```python
def plotUpdate(first, second, title, firstLabel, secondLabel,  ylabel, xlim, ylim, loc = False):
    firstLine, = plt.plot([i for i in range(0,len(first))], first)
    secontLine, = plt.plot([i for i in range(0,len(second))], second)
    plt.xlabel('epoch')
    plt.ylim(ylim)
    plt.ylabel(ylabel)
    plt.title(title)
    if loc:
        plt.legend([firstLine, secontLine] [firstLabel, secondLabel], loc = (0.624,0))
    else:
        plt.legend([firstLine, secontLine], [firstLabel, secondLabel])
    plt.grid(True)
```

```python
plotUpdate([100, 10.6866, 11.1927, 9.2235, 8.6792, 7.2750, 6.1111, 6.2611, 5.9202, 5.7136],
           [100, 12.3792, 10.8379, 9.1064, 9.0443, 6.8364, 6.3481, 6.1579, 6.1692, 6.3350],
            'Loss over training', 'training set', 'test set', 'loss', (0,13))
```
